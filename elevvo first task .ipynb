```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\HP\Downloads\study_hours_scores_regression (1).csv")
data
```

```python
''' first we divide the data to x and y 
x => should be matrix 
y => should be vector 
'''
```

```python
x=data.iloc[:,[0]].values 
```

```python
y=data.iloc[:,1].values 
```

```python
from sklearn.model_selection import train_test_split
```

```python
''' we divide our data to train and test '''
```

```python
x_train ,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
regressor = LinearRegression()
```

```python
''' fit function is step to train data'''
```

```python
regressor.fit(x_train,y_train)
```

```python
''' testing step '''
```

```python
ypred=regressor.predict(x_test)
```

```python
# evaluation of model 
from sklearn.metrics import mean_absolute_percentage_error
print (mean_absolute_percentage_error(y_test , ypred))
```

```python
print(f"MAE : {regressor.predict([[3]])}")
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, ypred)
```

```python
print(f" RÂ² Score: {r2:.4f}")
```

```python
plt.scatter(x_train ,y_train)
plt.plot(x_train ,regressor.predict(x_train),color='red')
plt.xlabel("Hours of study")
plt.ylabel("Degree")
```
